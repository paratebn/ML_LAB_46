```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn import linear_model

# Assuming 'data' is your DataFrame
X_train, X_test, y_train, y_test = train_test_split(
    data[['5.1', '3.5', '1.4', '0.2']],
    data['Iris-setosa'],
    test_size=0.3
)

# Remove rows with NaN in y_train
train_data = pd.concat([X_train, y_train], axis=1)  # Combine X_train and y_train
train_data = train_data.dropna(subset=[y_train.name])  # Drop rows with NaN in y_train column

# Update X_train and y_train
X_train = train_data[['5.1', '3.5', '1.4', '0.2']]
y_train = train_data[y_train.name]

mymodel = linear_model.LogisticRegression(max_iter=120)
mymodel.fit(X_train, y_train)
```

```python
mymodel.predict(X_test)
```

```python
X_test
```

```python
test_data = pd.concat([X_test, y_test], axis=1) # Combine X_test and y_test
test_data = test_data.dropna(subset=[y_test.name]) # Drop rows with NaN in y_test column
X_test = test_data[['5.1', '3.5', '1.4', '0.2']]
y_test = test_data[y_test.name]

mymodel = linear_model.LogisticRegression(max_iter=120)
mymodel.fit(X_train, y_train)
```

```python
mymodel.score(X_test, y_test)
```

```python
predicted_output = mymodel.predict(X_test)
predicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_output)
cm
```
